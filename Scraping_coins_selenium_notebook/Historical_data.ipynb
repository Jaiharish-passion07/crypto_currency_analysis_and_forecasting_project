```python
#Import libraires
import pandas as pd
from selenium.webdriver.common.by import By
import time
from selenium import webdriver
from datetime import datetime
from selenium.webdriver.chrome.service import Service
```

```python
#path for chrome webdriver
path=Service("C:\selenium drivers\chromedriver.exe")
op = webdriver.ChromeOptions()
```

####  *Bitcoin Historical data only available from April 28 2013 to till date in the Coin Marketcap website*

```python
date_1 = '1/1/2022'
date_2 = '16/6/2022'
start = datetime.strptime(date_1, "%d/%m/%Y")
end =   datetime.strptime(date_2, "%d/%m/%Y")
diff = (end.year - start.year) * 12 + (end.month  - start.month )
no_of_month=diff#It decides number of time to click load more button in coinmarketcap historical page 
delta = end.date() - start.date()
no_of_records=int(delta.days)#It iterate the each days using no of records var
```

#### *You can also customize your start and end date of time period by changing above code and choose your favorite coin below to scrap the historical data*

```python
select_coins=['bitcoin','ethereum','bnb','solana','polkadot-new','cardano']
coin_choosen=input(f"Select Number to choose coin \n Bitcoin(1) \n Ethereum(2) \n BNB(3) \n Solana(4) \n Polkadot(5) \n Cardano(6)")
coin_chosen_by_user=select_coins[int(coin_choosen)-1]
print(coin_chosen_by_user)
```

#### *Scraping the Data from Coin Marketcap*

```python
#main code here
try:
    date_list=[]
    open_price_list=[]
    high_price_list=[]
    low_price_list=[]
    close_price_list=[]
    volume_list=[]
    market_cap_list=[]
    #URL path
    driver = webdriver.Chrome(service=path, options=op)
    # For maximizing window
    driver.maximize_window()
    # Implicit wait apply for all elements to wait for 5 seconds to load
    driver.implicitly_wait(5)
    driver.get("https://coinmarketcap.com/currencies/bitcoin/historical-data/")
    time.sleep(3)
    driver.find_element(by=By.XPATH,value="//*[@id='cmc-cookie-policy-banner']/div[2]").click()
    for j in range(1,no_of_month):
        time.sleep(3)
        driver.find_element(by=By.XPATH,value='//button[normalize-space()="Load More"]').click()
        
    for i in range(1,no_of_records+2):
        
        #Accessing coin date, open_price,close_price_low_price_high_price,,volume_24hr_change, Mkt_Cap using the element xpath

        date=driver.find_element(by=By.XPATH,value="//*[@id='__next']/div[1]/div[1]/div[2]/div/div[3]/div/div/div[1]/div[2]/table/tbody/tr["+str(i)+"]/td[1]")
        open_price=driver.find_element(by=By.XPATH,value="//*[@id='__next']/div[1]/div[1]/div[2]/div/div[3]/div/div/div[1]/div[2]/table/tbody/tr["+str(i)+"]/td[2]")
        high_price=driver.find_element(by=By.XPATH,value="//*[@id='__next']/div[1]/div[1]/div[2]/div/div[3]/div/div/div[1]/div[2]/table/tbody/tr["+str(i)+"]/td[3]")
        low_price=driver.find_element(by=By.XPATH,value="//*[@id='__next']/div[1]/div[1]/div[2]/div/div[3]/div/div/div[1]/div[2]/table/tbody/tr["+str(i)+"]/td[4]")
        close_price=driver.find_element(by=By.XPATH,value="//*[@id='__next']/div[1]/div[1]/div[2]/div/div[3]/div/div/div[1]/div[2]/table/tbody/tr["+str(i)+"]/td[5]")
        volume=driver.find_element(by=By.XPATH,value="//*[@id='__next']/div[1]/div[1]/div[2]/div/div[3]/div/div/div[1]/div[2]/table/tbody/tr["+str(i)+"]/td[6]")
        mak_cap=driver.find_element(by=By.XPATH,value="//*[@id='__next']/div[1]/div[1]/div[2]/div/div[3]/div/div/div[1]/div[2]/table/tbody/tr["+str(i)+"]/td[7]")
        
        #Appending all Data into respective records
        date_list.append(date.text.strip())
        open_price_list.append(open_price.text.strip())
        high_price_list.append(high_price.text.strip())
        low_price_list.append(low_price.text.strip())
        close_price_list.append(close_price.text.strip())
        volume_list.append(volume.text.strip())
        market_cap_list.append(mak_cap.text.strip())

except:
    #setting time to close after 2 sec
    time.sleep(3)
    driver.close()
finally:
    time.sleep(3)
    driver.close()
```

#### *Creating Dataframe from scraped results*

```python
#Create Pandas DataFame
bitcoin_df1=pd.DataFrame({"Date":date_list,"Open Price":open_price_list,"High Price":high_price_list,"Low Price":low_price_list,"Close Price":close_price_list,"Volume_24hr":volume_list,"Market Cap":market_cap_list})
```
